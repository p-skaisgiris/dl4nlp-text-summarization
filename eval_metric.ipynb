```python
import pandas as pd 
pd.set_option('display.float_format', '{:.4f}'.format)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

path = "human_judge_summ_metrics_2023_10_14_22_45.csv"

df = pd.read_csv(path)

# drop nan based on qags
drop_nan = df.dropna(subset=['qags'])

print(drop_nan.shape)
df_select = drop_nan[['human_judgement', 'qags', 'rouge', 'triples', 'bleurt', 'summac', 'ensemble']]

corr = df_select.corr()
heatmap = sns.heatmap(corr, annot=True, cmap="Blues", fmt='.1g')
```
